Lab02 part2 lab task is to visualize the data set or the data using Matplotlib

```python
import matplotlib.pyplot as plt
people = ["Sushruth","Gamma","Aditya","Sathwik"]
weight = [61,70,40,69]
age = [17,18,19,20]
height = [173,172,171,172]
plt.scatter(weight,height)
plt.title("Weight-Height Relation")
plt.ylabel("Height in cms")
plt.xlabel("weight in kgs")
plt.show()

plt.scatter(age,weight)
plt.title("Age-Weight Relation")
plt.xlabel("Age in years")
plt.ylabel("weight in kgs")
plt.show()
```

#Generate Bar Plot

```python
plt.figure(figsize = (12,3))
plt.title("People's weight in kgs",fontsize = 24,fontstyle='italic')
plt.bar(x=people,height=weight,width=0.6,color='red',alpha = 0.1)
plt.xlabel("People",fontsize=26)
plt.ylabel("Weight(in kgs)",fontsize=19)
plt.show()
```

#GENRATE HISTOGRAM

```python
plt.figure(figsize=(12,4))
plt.hist(weight,color='red',edgecolor='k',alpha=0.4,bins=9)

plt.title("histogram of patient weight",fontsize=15)
plt.xlabel("Weightinkgs",fontsize= 23)

plt.show()
```
